Создать базовый класс Customer в C#, который будет представлять информацию о клиентах или покупателях.

На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма.

В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

# Требования к базовому классу Customer:
## **• Атрибуты:** Идентификатор клиента (CustomerId), Имя (Name), Электронная почта (Email).
## **• Методы:**
### o GetFullName(): метод для получения полного имени клиента.
### o UpdateEmail(string newEmail): метод для обновления электронной почты клиента.
### o ViewProfile(): метод для просмотра профиля клиента.
Требования к производным классам:
1. VIPКлиент (VipCustomer): Должен содержать дополнительные атрибуты,
такие как Баланс лояльности (LoyaltyPoints). Метод ViewProfile() должен быть
переопределен для отображения дополнительной информации о VIPклиенте.

2. ОбычныйКлиент (RegularCustomer): Должен содержать дополнительные
атрибуты, такие как Дата регистрации (RegistrationDate).
Метод UpdateEmail() должен быть переопределен для добавления
информации о дате последнего обновления электронной почты.

3. ГрупповойКлиент (GroupCustomer) (если требуется третий класс): Должен
содержать дополнительные атрибуты, такие как Название группы
(GroupName). Метод GetFullName() должен быть переопределен для
отображения названия группы вместо имени клиента.

# Дополнительное задание
**Добавьте к сущестующим классам конструктора классов с использованием гетторов и сетторов и реализуйте взаимодействие объектов между собой**

```
public class Customer 
{
    //ATRIBUTES
    private int _customerid; // customerid
    public int CustomerId 
    {
        get{return _customerid;}
        set{_customerid = value;}
    }

    private string _name;
    public string Name 
    {
        get {return _name;} 
        set{_name = value;}
    }

    private string _email;
    public string Email 
    {
        get {return _email;} 
        set{_email = value;}
    }

    public static int CustomerCount {get;private set;}

    public Customer(int customerid, string name, string email)
    {
        CustomerCount++;
        CustomerId = customerid;
        Name = name;
        Email = email;
    }

    //METHODS
    public virtual void GetFullName()
    {
        Console.WriteLine($"Имя клиента - {Name}");
    }

    public virtual void UpdateEmail(string NewEmail)
    {
        Console.WriteLine($"Старая почта {Email} обновлена на новую - {NewEmail}");
        Email = NewEmail;

    }

    public virtual void ViewProfile()
    {
        Console.WriteLine($"\nПрофиль клиента:\nАйди клиента {CustomerId}, \nИмя клиента {Name}, \nПочта клиента {Email}.");
    }

    
}

public class VipCustomer : Customer
{
    private int _loyalityPoints;
    public int LoyalityPoints 
    {
        get{return _loyalityPoints;}
        set{_loyalityPoints = value;}
    }

    public VipCustomer(int customerid, string name, string email) : base(customerid, name, email)
    {
        CustomerId = customerid;
        Name = name;
        Email = email;
    }

    public override void ViewProfile ()
    {
        Console.WriteLine($"\nПрофиль VIP клиента:\nАйди клиента {CustomerId}, \nИмя клиента {Name}, \nПочта клиента {Email}\nБаланс лояльности: {LoyalityPoints}.");
    }


}

public class RegularCustomer : Customer
{
    private string  _registrationdate;
    public string  RegistrationDate 
    {
        get{return _registrationdate;}
        set{_registrationdate = value;}
    }

    public RegularCustomer (int customerid, string name, string email) : base(customerid, name, email)
    {
        CustomerId = customerid;
        Name = name;
        Email = email;
    }

   public override void UpdateEmail(string NewEmail)
    {
        Console.WriteLine($"\nДата последнего изменения почты: {RegistrationDate}\nСтарая почта {Email} обновлена на новую - {NewEmail}");
        Email = NewEmail;
    }

        public override void ViewProfile()
    {
        Console.WriteLine($"\nПрофиль клиента:\nАйди клиента {CustomerId}, \nИмя клиента {Name}, \nПочта клиента {Email}\nДата регистрации аккаунта: {RegistrationDate}.");
    }
}

public class GroupCustomer : Customer
{
    private string _groupName;
    public string GroupName 
    {
        get{return _groupName;}
        set{_groupName = value;}
    }

    public GroupCustomer (int customerid, string name, string email) : base(customerid, name, email)
    {
        CustomerId = customerid;
        Name = name;
        Email = email;
    }

    public override void GetFullName()
    {
        Console.WriteLine($"\nИмя группы: {GroupName}");
    }
}

// MAIN
Customer customer = new Customer(156, "Bob Jhonson", "Bobs@email.com");
VipCustomer vipcustomer = new VipCustomer(200, "Boris Borisovich", "BorisBorisovich@email.com");
RegularCustomer regularcustomer = new RegularCustomer(112, "Обычный заказчик", "REGULAR_CUSTOMER@email.com");

//Customer
Console.WriteLine("\n!Customer");
customer.GetFullName();
customer.UpdateEmail("NewBobs@email.com");
customer.ViewProfile();

//VipCustomer
Console.WriteLine("\n!VipCustomer");
vipcustomer.LoyalityPoints = 98738242;
vipcustomer.ViewProfile();

//RegularCustomer
Console.WriteLine("\n!RegularCustomer");
regularcustomer.RegistrationDate = "02.10.2007";
regularcustomer.UpdateEmail("NEW_REGULAR_CUSTOMER@email.com");
regularcustomer.ViewProfile();

//GroupCustomer
Console.WriteLine("\n!GroupCustomer");
GroupCustomer groupcustomer = new GroupCustomer(130, "Олег", "Oleg@Email.com");
groupcustomer.GroupName = "Легкая атлетика";
groupcustomer.GetFullName();
```
