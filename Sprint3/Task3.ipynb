# Требования к базовому классу Customer:
## **• Атрибуты:** Идентификатор клиента (CustomerId), Имя (Name), Электронная почта (Email).
## **• Методы:**
### o GetFullName(): метод для получения полного имени клиента.
### o UpdateEmail(string newEmail): метод для обновления электронной почты клиента.
### o ViewProfile(): метод для просмотра профиля клиента.
Требования к производным классам:
1. VIPКлиент (VipCustomer): Должен содержать дополнительные атрибуты,
такие как Баланс лояльности (LoyaltyPoints). Метод ViewProfile() должен быть
переопределен для отображения дополнительной информации о VIPклиенте.

2. ОбычныйКлиент (RegularCustomer): Должен содержать дополнительные
атрибуты, такие как Дата регистрации (RegistrationDate).
Метод UpdateEmail() должен быть переопределен для добавления
информации о дате последнего обновления электронной почты.

3. ГрупповойКлиент (GroupCustomer) (если требуется третий класс): Должен
содержать дополнительные атрибуты, такие как Название группы
(GroupName). Метод GetFullName() должен быть переопределен для
отображения названия группы вместо имени клиента.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```
using System;
using System.Collections.Generic;

public interface ICustomer
{
    string CustomerType {get; set;} // vip regular
    int Cost {get; set;} // Abonement costs
}

public interface IRegularCustomer
{
    string RegistrationDate {get; set;}
}

public interface IGroupCustomer
{
    string GroupName {get; set;}
}

public interface IVipCustomer
{
    int LoyalityPoints {get; set;}
}

//                                                                  MAIN CLASS
public class Customer : ICustomer
{
    public Random rnd = new Random();
    //ATRIBUTES

    public int CustomerId {get; set;}
    public string Name {get; set;}
    public string Email {get; set;}
    public string CustomerType {get; set;} 
    public Boolean IsMan; 
    public string Gender;
    public int Cost{get; set;}

    public static int CustomerCount {get; private set;}

    public Customer (string name, string email, Boolean isman)
    {
        CustomerCount++;
        CustomerType = "";
        CustomerId = rnd.Next(100,1000);
        IsMan = isman;
        Name = name;
        Email = email;
        Cost = 0;
        if (IsMan) {Gender = "Мужчина";} else{Gender = "Женщина";}
    }

    //METHODS
    public virtual void GetFullName()
    {
        Console.WriteLine($"Имя клиента - {Name}");
    }

    public virtual void UpdateEmail(string NewEmail)
    {
        Console.WriteLine($"Старая почта {Email} обновлена на новую - {NewEmail}");
        Email = NewEmail;
    }

    public virtual void ViewProfile()
    {
        Console.WriteLine($"Профиль клиента:\nАйди клиента {CustomerId}, \nТип клиента {CustomerType} \nПол клиента {Gender} \nИмя клиента {Name}, \nПочта клиента {Email}. \nСтоимость абонемента {Cost}");
    }

    public virtual void WhoInvited (string FriendName)
    {
        Console.WriteLine($"{Name}, Вас пригласил: {FriendName}");
    }

    public virtual void WhoInvited (int Count)
    {
        Console.WriteLine($"{Name}, если ваш друг пригласит ещё {Count} человек, то вы улучшите свой статус пользователя. Сейчас ваш статус: {CustomerType}");
    }
}

//                                                                  REGULAR CUSTOMER
public class RegularCustomer : Customer, IRegularCustomer
{
    public string RegistrationDate {get; set;}

    public RegularCustomer (string name, string email, Boolean isman) : base(name, email, isman)
    {
        CustomerType = "RegularCustomer";
        RegistrationDate = $"{rnd.Next(1,31)},{rnd.Next(1,12)},{rnd.Next(2000,2024)}";
        Cost = 1000;
    }

   public override void UpdateEmail(string NewEmail)
    {
        Console.WriteLine($"\nДата последнего изменения почты: {RegistrationDate}\nСтарая почта {Email} обновлена на новую - {NewEmail}");
        Email = NewEmail;
    }

    public override void ViewProfile()
    {
        Console.WriteLine($"Профиль REGULAR клиента:\nАйди клиента {CustomerId}, \nТип клиента {CustomerType} \nПол клиента {Gender} \nИмя клиента {Name}, \nПочта клиента {Email}. \nСтоимость абонемента {Cost} \nДата регистрации{RegistrationDate}");
    }

    public override void WhoInvited (string FriendName)
    {
        Console.WriteLine($"{Name}, Вас пригласил: {FriendName}");
    }

    public override void WhoInvited (int Count)
    {
        Console.WriteLine($"{Name}, если ваш друг пригласит ещё {Count} человек, то вы улучшите свой статус пользователя. Сейчас ваш статус: {CustomerType}");
    }
}


//                                                                  GROUP CUSTOMER
public class GroupCustomer : RegularCustomer, IGroupCustomer
{
    public string GroupName {get; set;}

    public GroupCustomer(string name, string email, Boolean isman, string groupname) : base(name, email, isman)
    {
        CustomerType = "GroupCustomer";
        GroupName = groupname;
        Cost = 800;
    }

    public override void GetFullName()
    {
        Console.WriteLine($"Имя группы: {GroupName}");
    }

    public override void ViewProfile()
    {
        Console.WriteLine($"Профиль GROUP клиента:\nАйди клиента {CustomerId}, \nТип клиента {CustomerType} \nПол клиента {Gender} \nИмя клиента {Name}, \nПочта клиента {Email}. \nСтоимость абонемента {Cost} \nДата регистрации {RegistrationDate}");
    }

        public override void WhoInvited (string FriendName)
    {
        Console.WriteLine($"{Name}, Вас пригласил: {FriendName}");
    }

    public override void WhoInvited (int Count)
    {
        Console.WriteLine($"{Name}, если ваш друг пригласит ещё {Count} человек, то вы улучшите свой статус пользователя. Сейчас ваш статус: {CustomerType}");
    }
}

//                                                                  VIP CUSTOMER
public class VipCustomer : RegularCustomer, IVipCustomer
{
    public int LoyalityPoints {get; set;}

    public VipCustomer (string name, string email, Boolean isman) : base(name, email, isman)
    {
        CustomerType = "VIP";
        LoyalityPoints = rnd.Next(1000,5000);
        Cost = 1600;
    }

    public override void ViewProfile ()
    {
        Console.WriteLine($"Профиль VIP клиента:\nАйди клиента {CustomerId}, \nТип клиента {CustomerType} \nПол клиента {Gender} \nИмя клиента {Name}, \nПочта клиента {Email}. \nСтоимость абонемента {Cost} \nДата регистрации {RegistrationDate} \nОчки лояльности ВИП клиента {LoyalityPoints}");
    }

    public override void WhoInvited (string FriendName)
    {
        Console.WriteLine($"{Name}, Вас пригласил: {FriendName}");
    }

    public override void WhoInvited (int Count)
    {
        Console.WriteLine($"{Name}, если ваш друг пригласит ещё {Count} человек, то вы улучшите свой статус пользователя. Сейчас ваш статус: {CustomerType}");
    }
}

public class CustomerCollection<T> where T : Customer
{
    private List<T> _customer = new List<T>();

    public void Add(T cu)
    {
        _customer.Add(cu);
    }

    public void Remove(T cu)
    {
        _customer.Remove(cu);
    }

    public void WhoInvited()
    {
        foreach (var cu in _customer)
        {
            cu.WhoInvited("Жора");
            cu.WhoInvited(14);
            Console.WriteLine();
        }
    }

    public void GetCustomerDetails()
    {
        foreach (var cu in _customer)
        {
            cu.GetFullName();
            cu.ViewProfile();
            Console.WriteLine();
        }
    }

}

// MAIN
CustomerCollection<Customer> customer = new CustomerCollection<Customer>();
customer.Add(new VipCustomer("Bob", "Bobs@email.com", true));
customer.Add(new RegularCustomer("Anna", "AnnaAnn@email.com", false));
customer.Add(new GroupCustomer("John", "Group_Customer@email.com", true, "FITNESS"));

customer.GetCustomerDetails();

/*
//RegularCustomer
Console.WriteLine("REGULAR CUSTOMER");
regularcustomer.GetFullName();*/
regularcustomer.UpdateEmail("NewAnna@email.com");
Console.WriteLine();
//regularcustomer.ViewProfile();

//VipCustomer
//Console.WriteLine("\nVipCustomer");
//Console.WriteLine();
//vipcustomer.ViewProfile();

/*GroupCustomer
Console.WriteLine("\nGroupCustomer");
groupcustomer.GetFullName();
Console.WriteLine();
groupcustomer.ViewProfile();*/

customer.GetCustomerDetails();
```
