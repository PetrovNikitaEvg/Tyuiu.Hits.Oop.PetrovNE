```csharp
public class Product
{
    private string name;
    public string Name 
    {
        get {return name;} 
        set{name = value;}
    }

    private decimal price {get; set;} // цена

    private int quantily; // количетсов товара
    public int Quantily 
    {
        get {return quantily;} 
        set{quantily = value;}
    }

    protected string category {get; set;} // категория товара
    internal string description {get; set;} // описание товара

    public Product (string _name, decimal Price, string Category, string Description, int _quantily)
    {
        name = _name;
        price = Price;
        category = Category;
        description = Description;
        quantily = _quantily;
    }

    public decimal GetTotalPrice() // GET COST ALL PRODUCTS
    {
        return price * quantily;
    }

    public void UpdateQuantity(int amount) // Update product amount
    {
        Console.WriteLine($"Было всего товаров {quantily} стало {quantily + amount}.");
        quantily += amount;
    }

    public string GetProductInfo()
    {
        return $"Название товара {name} \nСтоимость одной еденицы товара: {price} \nКатегория товара: {category}\nКоличество товара: {quantily}\nОписание: {description}";
    }

}

public class Seller
{
    private string name;
    public string Name 
    {
        get {return name;} 
        set{name = value;}
    }

    private string emloyeeid{get; set;}
    protected decimal salary{get; set;} // зп
    internal string contactinfo{get; set;}
    private string SellerProductsName = "";

    public Seller (string _name, string EmloyeeId, decimal Salary, string ContactInfo)
    {
        name = _name;
        emloyeeid = EmloyeeId;
        salary = Salary;
        contactinfo = ContactInfo;
    }

    public void AddPdroduct(Product product)
    {
        SellerProductsName += " " + product.Name;
    }

    public void SellProduct(Product product, int quantity) // уменьшает количество
    {
        Console.WriteLine($"Продавец продал {quantity} товара с именем {product.Name}. Всего было {product.Quantily} стало {product.Quantily - quantity}");
        product.Quantily -= quantity;
    }

    public string GetSellerInfo() // Возвращает информацио о продавце
    {
        return $"Имя продавца {name} \nАйди продавца: {emloyeeid} \nЗарплата продавца: {salary}\nКонтакты продавца: {contactinfo} \nСписок товаров продавца{SellerProductsName}";
    }
}

class Store
{
    private string storename;
    private string location;
    public string storehours;
    private string SellersList = "";
    public string productslist = "";

    public Store (string StoreName, string Location, string StoreHours)
    {
        storename = StoreName;
        location = Location;
        storehours = StoreHours;
    }

    public void AddSeller(Seller seller)
    {
        SellersList = " " + seller.Name;
    }

    public void ListProducts()
    {
        Console.WriteLine(productslist);
    }

    public string GetStoreInfo()
    {
        return $"Имя магазина: {storename}\nЛокация магазина: {location}\nВремя работы магазина: {storehours}\nСписок продавцов: {SellersList}\nСписок всех товаров магазина: {productslist}";
    }
}

Store store = new Store("MyStore", "NewYork", "24");
Product product =  new Product("Стул", 17952.143m, "Мебель", "КРУТОЙ СТУЛ, УДОБНЫЙ", 8);
store.productslist += " " + product.Name;
Seller seller = new Seller("Bob", "001", 12331.10m, "VK TELEGRAM");

store.AddSeller(seller);
store.ListProducts();
Console.WriteLine(store.GetStoreInfo());

Console.WriteLine($"Общая стоимость товаров {"СТУЛ"} {product.GetTotalPrice()}");
product.UpdateQuantity(2); //+к товару
Console.WriteLine();
Console.WriteLine(product.GetProductInfo());

seller.AddPdroduct(product);
Console.WriteLine();
Console.WriteLine(seller.GetSellerInfo());

seller.SellProduct(product, 1);

Console.Write($"Количество товара {product.Name} = {product.Quantily}");
```
